```python
import nltk
import re
import pandas as pd
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
```

```python
data = pd.read_csv('tripadvisor_hotel_reviews.csv')
```

```python
data
```

```python
data.info()
```

```python
data.head()
```

```python
data['Review'][0]
```

```python
data['review_lowercase'] = data['Review'].str.lower()
```

```python
data.head()
```

```python
en_stopwords = stopwords.words('english')
```

```python
en_stopwords.remove('not')
```

```python
data['review_no_stopwords'] = data['review_lowercase'].apply(lambda x: ' '.join(word for word in x.split() if word not in (en_stopwords)))
```

```python
data.head()
```

```python
data['review_no_stopwords'][0]
```

```python
data['review_no_stopwords_no_punch'] = data.apply(lambda x: re.sub(r"[*]", "star", x['review_no_stopwords']), axis=1)
```

```python
data.head()
```

```python
data['review_no_stopwords_no_punch'] = data.apply(lambda x: re.sub(r"[^\w\s]", "", x['review_no_stopwords_no_punch']), axis=1)
```

```python
data['review_no_stopwords_no_punch'][0]
```

```python
data['tokenize'] = data.apply(lambda x: word_tokenize(x['review_no_stopwords_no_punch']), axis=1)
```

```python
data.head()
```

```python
ps = PorterStemmer()
```

```python
data['stemmed'] = data['tokenize'].apply(lambda tokens: [ps.stem(token) for token in tokens])
```

```python
data.head()
```

```python
lemmitizer = WordNetLemmatizer()
```

```python
data['lemmitized'] = data['tokenize'].apply(lambda tokens: [lemmatizer.lemmatize(token) for token in tokens])
```

```python
data.head()
```

```python
data['lemmitized']
```

```python
tokens_clean = sum(data['lemmitized'], [])
```

```python
data.head()
```

```python
unigrams = (pd.Series(nltk.ngrams(tokens_clean, 1)).value_counts())
```
